```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
df = pd.read_csv('Salary Data.csv')
print(df)
```

```python
print(df.isnull().sum())
```

```python
new_df = df.dropna()
new_df
```

```python
new_df.isnull().sum()
```

```python
new_df = new_df.drop(['Gender', 'Education Level', 'Job Title'], axis=1)
```

```python
new_df.to_csv('new sales data.csv')
```

```python
new_sales = pd.read_csv('new sales data.csv')
new_sales
```

```python
new_sales.isnull().sum()
```

```python
x = new_sales[['Years of Experience']]
y = new_sales['Salary']

x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Root Mean Squared Error:",root_mean_squared_error(y_test,y_pred))
print("R^2 Score:",r2_score(y_test,y_pred))
```

```python
plt.scatter(new_sales['Years of Experience'],new_sales['Salary'],s=80)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Classification data")
plt.grid(True)
plt.show()
```
