```python
import numpy as np
import tensorflow as tf
```

```python
temperature = np.array([20,22,24,26,28])
icecream = np.array([200,250,300,350,400])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001),loss='mean_squared_error')

model.fit(temperature,icecream,epochs=200,verbose=0)

print(model.predict(np.array([30])))
```
