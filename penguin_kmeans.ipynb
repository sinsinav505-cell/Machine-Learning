```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = pd.read_csv('penguins.csv')
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df
```

```python
x = df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']].dropna()

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(x)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10,6))
plt.scatter(x['culmen_length_mm'], x['culmen_depth_mm'], c=labels)
plt.scatter(centroids[:,0],centroids[:,1], c='red', marker='X', s=200, label='centroids')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Depth (mm)')
plt.title('K-Means Clustering (2 Features)')
plt.show()
```

```python
import numpy as np

sample = np.array([[39.1,18.7,181,3750]])

sample_cluster = kmeans.predict(sample)
print("sample belongs to cluster:",sample_cluster)

plt.scatter(x.iloc[:,0], x.iloc[:,1], c=labels)
plt.scatter(sample[:,0],sample[:,1], s=200, marker='*', color='red', label='Sample')
plt.xlabel("Feature 1(Sepal length)")
plt.ylabel("Feature 2(Sepal width)")
plt.title("K-Means New sample prediction")

plt.legend()
plt.show()
```
