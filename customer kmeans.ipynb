```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
x = df[['AnnualIncome','SpendingScore']]
kmeans = KMeans(n_clusters=5,random_state=42) 
kmeans.fit(x)

y_pred = kmeans.labels_
print("Cluster labels:",y_pred)

plt.figure(figsize=(10,6))

for i in range(5):  
    plt.scatter(
        x[y_pred == i]['AnnualIncome'],
        x[y_pred == i]['SpendingScore'],
        label=f"Cluster {i}"
    )

plt.xlabel("AnnualIncome")
plt.ylabel("SpendingScore")
plt.title("K-Means Clustering Result")
plt.legend()
plt.show()
```

```python
sample = np.array([[20,40]])

sample_cluster = kmeans.predict(sample)
print("sample belongs to cluster:",sample_cluster)

plt.figure(figsize=(10,6))

for i in range(5):  
    plt.scatter(
        x[y_pred == i]['AnnualIncome'],
        x[y_pred == i]['SpendingScore'],
        label=f"Cluster {i}"
    )
plt.scatter(sample[:,0],sample[:,1], s=200, marker='*', color='red', label='Sample')
plt.xlabel("AnnualIncome")
plt.ylabel("SpendingScore")
plt.title("K-Means Clustering Result")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_scaled)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels)
print(centroids)



plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels,  edgecolors='k',s=50)
plt.scatter(centroids[:,0],centroids[:,1], c='red', marker='X', s=200, label='centroids')
plt.xlabel("Anual income")
plt.ylabel("Spending Score")
plt.title("K-Means clustering")

plt.legend()
plt.show()
```
