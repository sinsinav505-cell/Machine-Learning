```python

class BankAccount:
    
    def __init__(self):
        self.balance = 0
        self.interest_rate = 0.02

    def deposit_amount(self,deposit):
        self.balance += deposit
        print("Deposited amount:",deposit,"Total  balance:",self.balance)

    def withdraw_amount(self,withdraw):
        if withdraw>self.balance:
            print("Insufficient balance \n balance = ",self.balance)
        else:
            self.balance -= withdraw
            print("Withdrawed amount:",withdraw,"balance:",self.balance)

    def balance_amount(self):
        print("Balance :",self.balance)

    
    def add_interest_bank(self):
            interest = self.balance * self.interest_rate
            self.balance += interest
            print("BankAccount Interest (2%):", interest)
            print("New Balance:",self.balance)
            return interest




class SavingsAccount(BankAccount):

    def __init__(self):
        super().__init__()
        self.savings_rate = 0.03  

    def add_interest_savings(self):
        
        savings_interest = self.balance * self.savings_rate
        self.balance += savings_interest
        print("SavingsAccount Interest (3%):", savings_interest)

        print("New Balance:", self.balance)
        



bank = BankAccount()
savings = SavingsAccount()


def choose_account():
    print("\nSelect Account:")
    print("1. BankAccount")
    print("2. SavingsAccount")
    acc = int(input("Enter choice: "))
    if acc == 1:
        return bank
    elif acc == 2:
        return savings
    else:
        print("Invalid choice!")
        return None
    


def operator():
    while True:
        print("\n 1. Deposit Amount " 
        "\n 2. Withdraw amount " 
        "\n 3. Balance" 
        " \n 4 . Add Interest " 
        "\n 5. Exit")

        op = int(input("Enter the operation:"))

        if op in [1, 2, 3, 4]:
            account = choose_account()
            if account is None:
                continue
        
        if op == 1:
            deposit = int(input("Enter the amount to be deposited:"))
            account.deposit_amount(deposit)
        elif op == 2:
            withdraw = int(input("Enter the amount to be withdrawed:"))
            account.withdraw_amount(withdraw)

        elif op == 3:
            account.balance_amount()

        elif op == 4:
            print("\nChoose Account Type:")
            print("1. BankAccount (2%)")
            print("2. SavingsAccount (3%)")
            acc = int(input("Enter choice: "))

            if acc == 1:
                account.add_interest_bank()
            elif acc == 2:
                savings.add_interest_savings()
            else:
                print("Invalid account type!")

        elif op == 5:
            print("Exiting. Thank you!")
            break
        
        else:
            print("Invalid option. Try again.")
        
operator()
```
