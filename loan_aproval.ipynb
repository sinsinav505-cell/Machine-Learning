```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data ={
    'income':[30,40,50,60,70,35],
    'credit_score':[550,600,650,700,750,580],
    'approved':[0,0,1,1,1,0]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
x = df[['income','credit_score']]
y = df['approved']
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y ,test_size=0.2,random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accurracy score:",accuracy_score(y_test,y_pred))
```

```python
new_loan = [[90,850]]
prediiction = model.predict(new_loan)
print("passed if yes[1] ,no[0]:",prediiction[0])
```
