```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
data = {
    'tv_ads':[10,12,15,20,22,25],
    'radio_ads':[5,8,10,12,16,18],
    'sales':[120,150,170,200,230,260]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
x=df[["tv_ads"]]
y=df["sales"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y,test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Root Mean Squared Error:",root_mean_squared_error(y_test,y_pred))
print("R^2 score:",r2_score(y_test,y_pred))
```

```python
#radio

x1=df[["radio_ads"]]
y1=df["sales"]

x_train1 , x_test1 , y_train1 , y_test1 = train_test_split(x1, y1,test_size=0.2, random_state=42)

model1 = LinearRegression()
model1.fit(x_train1,y_train1)

y_pred1 = model1.predict(x_test1)

print("Root Mean Squared Error:",root_mean_squared_error(y_test1,y_pred1))
print("R^2 score:",r2_score(y_test1,y_pred1))
```

```python
plt.figure(figsize=(10,5))

# Plot 1
plt.subplot(1,2,1)
plt.scatter(x1, y1,color='blue')
plt.plot(x1, model1.predict(x1),color='orange')
plt.xlabel('Radio_ads')
plt.ylabel('Sales')
plt.title("Radio_ads vs Sales")

# Plot 2
plt.subplot(1,2,2)
plt.scatter(x, y,color='red')
plt.plot(x, model.predict(x),color='green')
plt.xlabel('Tv_ads')
plt.ylabel('Sales')
plt.title("Tv_ads vs Sales")

plt.show()
```
