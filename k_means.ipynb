```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = load_iris()  #load dataset
x = data.data
```

```python
kmeans = KMeans(n_clusters=3,random_state=42) #training
kmeans.fit(x)

y_pred = kmeans.labels_
print("Cluster labels:",y_pred)


plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means clustering result")

plt.legend([
    "Cluster 0= Iris-setosa",
    "Cluster 1= Iris-versicolor",
    "Cluster 2= Iris-verginica"
])
```

```python
import numpy as np

sample = np.array([[5.2,3.8,1.6,0.2]])

sample_cluster = kmeans.predict(sample)
print("sample belongs to cluster:",sample_cluster)

plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(sample[:,0],sample[:,1], s=200, marker='*', color='red', label='Sample')
plt.xlabel("Feature 1(Sepal length)")
plt.ylabel("Feature 2(Sepal width)")
plt.title("K-Means New sample prediction")

plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

iris = load_iris()
x = iris.data[:,:2]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_scaled)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels)
print(centroids)



plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels, cmap='viridis', edgecolors='k',s=50)
plt.scatter(centroids[:,0],centroids[:,1], c='red', marker='X', s=200, label='centroids')
plt.xlabel("Sepal length(standardized)")
plt.ylabel("Sepal width(standardized)")
plt.title("K-Means clustering iris dataset")

plt.legend()
plt.show()
```
